# Q5 - Find the top 3 Airlines which had the largest mean departure delay in the month of January (2 points)
Month of January has the id of 1. Consider only flights with the positive departure delay. 

```python
airline_depart_delay = df[(df['MONTH']  == 1) & (df['DEPARTURE_DELAY'] > 0)][['AIRLINE', 'DEPARTURE_DELAY']]
airline_depart_delay.groupby("AIRLINE").mean().sort_values(by='DEPARTURE_DELAY', ascending=False)[0:3]
```

## Q6 - What percentage of flights out of Austin (AUS) are canceled? (2 points)
The column field "CANCELLED" will be equal to one of the flight is cancelled. 

```python
all_flight = df[ (df["ORIGIN_AIRPORT"] == "AUS")]
canceled_flight = all_flight[df['CANCELLED'] == 1]

print(canceled_flight.size)
print(all_flight.size)

print("Percentage Canceled From Austin: " + str(canceled_flight.size / all_flight.size))
```
